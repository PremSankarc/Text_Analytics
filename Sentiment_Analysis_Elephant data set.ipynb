```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
stopword = set(stopwords.words('english'))

from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
```

```python
twet=pd.read_csv('tweets_ele2.csv', encoding='cp1252')
twet.head()
```

```python
new = pd.DataFrame({
        "text":twet['text'],
    })
new
```

```python
def preprocess(text):
    
    text=re.sub('RT','',text)
    
    #text=re.sub('...$','',text)
    
    # remove new lines
    text = text.replace('\n', ' ')
    text = text.replace('\r', ' ')
    # remove links
    text = re.sub('https?://\S+|www\.\S+', ' ', text)
    text = re.sub('http','',text)
    
    # remove hashtags at the end of text
    text = re.sub('#(?!(?:hashtag)\b)[\w-]+(?=(?:\s+#[\w-]+)*\s*$)', '',text)
    
    text=re.sub('<.+>','',text)
    
    # remove handles
    text = re.sub('@[\w]+', '',text)
    
    # remove punctuations
    punc ='''.?!,:;-_—[](){}'"`~|\/@#$%^&+=*¦'''
    for i in text:
        if i in punc:
            text = text.replace(i, '') 
            
    text=text.replace(" ' ","") 
                      
    # remove extra spaces
    re.sub("\s\s+", " ", text)
    
    # lower case
    text = text.strip().lower()
    
    # lemmatization
    text = [lemmatizer.lemmatize(word) for word in text.split(' ')]
    text=" ".join(text)
    
    # stopword removal
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    
    return text
```

```python
import string 
def remove_punctuations(text):
          return text.translate(string.punctuation)
```

```python
twet['text']= twet['text'].apply(preprocess)
#twet['text']= twet['text'].apply(remove_punctuations)
```

```python
for i in twet['text']:
    print(i)
    print('***')
```

```python
#finding sentiment score

polarity = [round(sentiment.polarity_scores(i)['compound'],2 ) for i in twet['text']]
twet['sentiment_score'] = polarity
```

```python
twet['score']=''
```

```python
for i in range(len(twet['sentiment_score'])):
    if twet['sentiment_score'][i]==0:
        twet['score'][i]=0
    elif twet['sentiment_score'][i]>0:
        twet['score'][i]=1
    else:
        twet['score'][i]=-1
```
